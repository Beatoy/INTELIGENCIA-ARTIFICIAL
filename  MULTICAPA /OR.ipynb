```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Datos OR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Definir el modelo de red neuronal
modelo = Sequential([
    Dense(4, activation='relu', input_dim=2),
    Dense(1, activation='sigmoid')
])

# Crear un optimizador Adam con una tasa de aprendizaje del 0.01
learning_rate = 0.01
adam_optimizer = Adam(learning_rate=learning_rate)

# Compilar el modelo
modelo.compile(
    optimizer=adam_optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
# Entrenar el modelo
history = modelo.fit(
    X, y,
    epochs=100,
    verbose=1,
)

# Graficar la evolución de la pérdida
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.title('Evolución de la pérdida (Loss)')
plt.xlabel('Época')
plt.ylabel('Binary Cross-Entropy')
plt.show()
```

```python
# Evaluar el modelo en el conjunto de entrenamiento
loss, accuracy = modelo.evaluate(X, y)
print(f'Loss: {loss}, Accuracy: {accuracy}')

# Predicción y comparación con valores reales
pred = (modelo.predict(X) > 0.5).astype(int)
for i, (inp, p) in enumerate(zip(X, pred)):
    print(f"{inp} -> {p[0]}")
```
