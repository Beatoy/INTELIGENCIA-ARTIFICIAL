```python
# Entradas para la neurona, así como están
#Entradas para la neurona
import numpy as np
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
# Salidas esperadas
Y = np.array([0, 0, 0, 1])

# Pesos iniciales, nada fuera de lo normal
W = [1, 1]

# Umbral para activar, básicamente la condición
umbral = 2

class McCullochPitts:
    def __init__(self, weights, umbral):
        """
            Constructor de la neurona MCP.
            Guarda las variables:
            weights -> vector de pesos iniciales
            umbral -> umbral para usar en la función de activación
        """
        # guardando los pesos así como vienen
        self.weights = weights

        # guardando el umbral
        self.umbral = umbral

    def predict(self, X):
        """
            Calcula la salida de la neurona para cada fila de entradas X.
        """
        # lista donde se guardan las salidas, sin complicaciones
        predictions = []

        # recorriendo cada fila de X, una por una
        for i in range(X.shape[0]):
            # calculando la suma ponderada: w*x
            weighted_sum = np.dot(self.weights, X[i])

            # imprimiendo para ver qué está pasando
            print(f"Input: {X[i]}, Weighted Sum: {weighted_sum}")

            # comparación con el umbral para activar o no
            if weighted_sum >= self.umbral:
                predictions.append(1)
            else:
                predictions.append(0)
        return predictions

# creando la instancia de la neurona MCP
mcp = McCullochPitts(W, umbral)

# obteniendo predicciones
predictions = mcp.predict(X)

# mostrando el resultado final
print("Predictions:", predictions)
```
