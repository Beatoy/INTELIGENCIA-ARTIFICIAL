```python
# OPERACION NOT

import numpy as np                                                      # importando numpy para hacer las operaciones sin complicarse

# Entradas para probar el NOT, solo usa un valor por vez
entradas = np.array([[0], [1]])                                         # aquí van las dos opciones posibles que puede recibir

# Peso que va a invertir la entrada
pesos = np.array([-1])                                                  # el peso negativo es lo que hace que se invierta la lógica

# Punto de activación
umbral = 0                                                              # si la suma llega a 0 o más, se considera como activado

# Donde se guardan los resultados finales
salidas = []                                                            # lista donde vamos metiendo cada salida calculada

for x in entradas:                                                      # recorriendo cada entrada que definimos
    suma = np.dot(x, pesos)                                             # multiplicación directa de la entrada con el peso
    print(f"Entrada: {x}, Suma ponderada: {suma}")                      # mostrando qué entrada se está usando y qué suma genera

    # Activación tipo escalón
    salida = 1 if suma >= umbral else 0                                 # revisa si se activa o no dependiendo del umbral
    salidas.append(salida)                                              # guardando la salida en la lista

# Ya mostrando todo junto
print("\nEntradas:\n", entradas)                                        # imprime todas las entradas que se probaron
print("Salidas NOT:", salidas)                                          # muestra lo que generó la neurona para cada caso
```
