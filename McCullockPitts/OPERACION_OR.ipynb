```python
import numpy as np                                         # usando numpy para manejar los arreglos sin complicaciones

# Entradas para la compuerta OR, todas las combinaciones posibles
entradas = np.array([[0,0], [0,1], [1,0], [1,1]])          # esta matriz contiene los pares binarios que se van a evaluar

# Pesos para cada entrada
pesos = np.array([1, 1])                                   # ambos pesos en 1 porque la OR solo necesita detectar si hay al menos un 1

# Nivel mínimo para activar
umbral = 1                                                 # si la suma llega a 1 o más, ya se considera que la OR da 1

# Aquí se guardan las salidas que vayamos obteniendo
salidas = []                                               # lista donde se va a ir acumulando cada salida

for x in entradas:                                         # recorriendo cada combinación de A y B
    suma = np.dot(x, pesos)                                # se hace la suma ponderada: simplemente la suma de los valores porque los pesos son 1
    print(f"Entrada: {x}, Suma ponderada: {suma}")         # mostrando qué valores entraron y cuánto dio la suma

    # activación tipo escalón
    salida = 1 if suma >= umbral else 0                    # si alcanza el umbral, la compuerta OR marca 1; si no, marca 0
    salidas.append(salida)                                 # la salida se guarda en la lista

# mostrando todo al final
print("\nEntradas:\n", entradas)                           # imprime las entradas completas usadas en la tabla
print("Salidas OR:", salidas)                              # imprime los resultados calculados de la operación OR
```
