```python
# Nota: ejecutar en un entorno con TensorFlow instalado (GPU opcional).

import numpy as np              # arreglos y operaciones numéricas
import tensorflow as tf         # base para construir y entrenar redes
from tensorflow.keras.models import Sequential     # modelo en forma de capas seguidas
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Softmax
from tensorflow.keras.optimizers import Adam       # optimizador más usado por su estabilidad
from tensorflow.keras.datasets import fashion_mnist   # dataset de ropa en 28x28
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Carga del dataset
# El dataset viene dividido automáticamente en entrenamiento y prueba.
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# 2) Exploración rápida
# Se muestra una imagen para confirmar que todo cargó bien.
index = 10
image = train_images[index]
label = train_labels[index]

print(image)   # matriz 28x28 con valores de gris

plt.figure(figsize=(3, 3))
plt.imshow(image, cmap=plt.cm.gray)   # se ve la prenda en escala de grises
plt.title(f'Etiqueta: {label} -> {class_names[label]}')
plt.axis('off')
plt.show()

# 3) Preprocesamiento
# Revisamos formas para confirmar que llegaron 60k imágenes de entrenamiento.
print('Forma train_images:', train_images.shape)
print('Forma test_images:', test_images.shape)

# Se normalizan los píxeles a rango [0,1], esto ayuda al entrenamiento.
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

# Las capas Conv2D necesitan una dimensión extra para canales (aquí 1 canal de gris).
train_images = np.expand_dims(train_images, -1)
test_images = np.expand_dims(test_images, -1)

# 4) Definición del modelo (CNN)
# Se arma una red básica con tres convoluciones y luego capas densas.
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),  # primera extracción de características
    MaxPooling2D((2, 2)),                                            # reduce tamaño a la mitad
    Conv2D(64, (3, 3), activation='relu'),                           # más filtros para detalles
    MaxPooling2D((2, 2)),                                            # otra reducción
    Conv2D(64, (3, 3), activation='relu'),                           # última conv
    Flatten(),                                                       # convierte el mapa en vector
    Dense(64, activation='relu'),                                    # capa oculta
    Dense(10)                                                        # logits finales para las 10 clases
])

model.summary()   # se imprime para revisar estructura y parámetros

# 5) Compilación
# Se usa Adam con lr estándar, y la pérdida acepta etiquetas enteras.
learning_rate = 0.001
adam_optimizer = Adam(learning_rate=learning_rate)

model.compile(
    optimizer=adam_optimizer,
    loss=SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

# 6) Entrenamiento
# Entrenamos por 10 épocas evaluando al mismo tiempo con el conjunto de prueba.
model.fit(train_images, train_labels, epochs=10,
          validation_data=(test_images, test_labels))

# 7) Evaluación final
# Se obtiene la pérdida y precisión del modelo ya entrenado.
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f'Pérdida en test: {test_loss:.4f} - Precisión en test: {test_acc:.4f}')

# 8) Predicciones y matriz de confusión
# Aquí se agrega Softmax para obtener probabilidades de cada clase.
probability_model = Sequential([model, Softmax()])

predictions = probability_model.predict(test_images)     # probabilidades
predicted_labels = np.argmax(predictions, axis=1)        # clase más probable

cm = confusion_matrix(test_labels, predicted_labels)

# Se grafica para ver qué tan bien predice cada categoría.
fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=class_names)
disp.plot(ax=ax, xticks_rotation='vertical')
plt.title('Matriz de confusión - Fashion MNIST')
plt.show()

# 9) Función para mostrar imágenes y predicciones
def plot_image(i, predictions_array, true_label, img):
    # Toma una imagen y muestra en el título qué predijo y cuál era la real.
    true_label, img = true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img.reshape((28, 28)), cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    color = 'blue' if predicted_label == true_label else 'red'   # azul si acierta

    plt.xlabel("{} {:2.0f}% ({})".format(
        class_names[predicted_label],
        100 * np.max(predictions_array),
        class_names[true_label]
    ), color=color)

# Se muestran varias predicciones al azar.
num_rows = 5
num_cols = 3
num_images = num_rows * num_cols

plt.figure(figsize=(2 * 2 * num_cols, 2 * num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 1)
    plot_image(i, predictions[i], test_labels, test_images)

plt.tight_layout()
plt.show()
```
