```python
# Perceptrón - Operación OR

import numpy as np

# Datos de entrenamiento: entradas y salidas esperadas para la OR
entradas = np.array([[0,0], [0,1], [1,0], [1,1]])
salidas_esperadas = np.array([0, 1, 1, 1])

# Inicializamos pesos y bias aleatorios (inicio del aprendizaje)
pesos = np.random.rand(2)
bias = np.random.rand(1)
tasa_aprendizaje = 0.1

# Función de activación tipo escalón: si la suma >= 0 devuelve 1, si no devuelve 0
def activacion(x):
    return 1 if x >= 0 else 0

# Entrenamiento simple: recorremos varias épocas y ajustamos pesos con la regla delta
for epoca in range(10):
    for i in range(len(entradas)):
        x = entradas[i]
        salida_deseada = salidas_esperadas[i]
        salida_obtenida = activacion(np.dot(x, pesos) + bias)

        # Regla de actualización del perceptrón: error * entrada * tasa
        error = salida_deseada - salida_obtenida
        pesos = pesos + tasa_aprendizaje * error * x
        bias = bias + tasa_aprendizaje * error

# Mostrar pesos y bias finales tras el entrenamiento
print("Pesos finales:", pesos)
print("Bias final:", bias)

# Prueba: aplicar el perceptrón entrenado a las entradas y mostrar resultados
print("\nResultados finales:")
for i in range(len(entradas)):
    x = entradas[i]
    salida = activacion(np.dot(x, pesos) + bias)
    print(f"Entrada: {x}, Salida: {salida}")
```
