```python
# McCulloch-Pitts - Operación NOT

import numpy as np

# Entradas binarias para la compuerta NOT (un solo bit por fila)
entradas = np.array([[0], [1]])

# Peso negativo; invierte el efecto de la entrada
pesos = np.array([-1])

# Umbral de activación
umbral = 0

# Lista donde se guardarán las salidas
salidas = []

for x in entradas:
    # Producto punto: w * x
    suma = np.dot(x, pesos)

    # Activación tipo escalón: determina 0 o 1
    salida = 1 if suma >= umbral else 0
    salidas.append(salida)

# Impresión final de entradas y salidas generadas
print("Entradas:\n", entradas)
print("Salidas NOT:", salidas)
```
